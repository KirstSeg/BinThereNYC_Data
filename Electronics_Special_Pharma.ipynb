```python
import requests

import pandas as pd
```

```python
base_url = 'https://data.cityofnewyork.us'
electronics = '/resource/wshr-5vic.json'
params = {
    "$select": "dropoff_sitename, address, borough, zipcode, latitude, longitude"
}

rE = requests.get(base_url + electronics, params = params)
```

```python
rE
```

```python
electronics = rE.json()
```

```python
electronics
```

```python
for i in electronics:
    i['Category'] = 'Electronics'
```

```python
import pandas as pd
data = electronics

dfE = pd.DataFrame(data)
dfE.index.name = 'Index'
dfE = dfE.rename(columns={'dropoff_sitename': 'Sitename', 'address':'Address', 'borough': 'Borough' , 'zipcode':'Zip' , 'latitude':'Latitude', 'longitude':'Longitude', 'acceptedMaterials':'Category'})

dfE.to_csv('electronics.csv', index=False)

dfE
```

```python
import folium

electronics_map = folium.Map(location=[dfE['Latitude'].astype(float).mean(), dfE['Longitude'].astype(float).mean()], zoom_start=10)
for Sitename, row in dfE.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Address'],
        tooltip=row['Sitename']
    ).add_to(electronics_map)
        
electronics_map
```

```python
base_url = 'https://data.cityofnewyork.us'
special = '/resource/242c-ru4i.json'
params = {
    "$select": "name, address, city, zip, point"
}

rSW = requests.get(base_url + special, params = params)
```

```python
rSW
```

```python
special_waste=rSW.json()
```

```python
special_waste
```

```python
for i in special_waste:
    i['Category'] = 'Special Waste'
    i['Longitude'] = i['point']['coordinates'][0]
    i['Latitude'] = i['point']['coordinates'][1]
    del i['point']
```

```python
data2 = special_waste

dfSW = pd.DataFrame(data2)
dfSW.index.name = 'Index'
dfSW = dfSW.rename(columns={'name': 'Sitename', 'address':'Address', 'city': 'Borough' , 'zip':'Zip'})

dfSW.to_csv('special_waste.csv', index=False)

dfSW
```

```python
special_waste_map = folium.Map(location=[dfSW['Latitude'].astype(float).mean(), dfSW['Longitude'].astype(float).mean()], zoom_start=10)
for Sitename, row in dfSW.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Address'],
        tooltip=row['Sitename']
    ).add_to(special_waste_map)
    
special_waste_map
```
