```python
batteries = response.json()
```

```python
batteries
```

```python
for i in batteries:
    i['addressZIP']=i['addressZIP'][:5]
    
batteries
```

```python
import csv

nyc_zip=[]

with open ("NYC-ZIP.csv","r") as f:
    datafile = csv.reader(f)
    for i in datafile:
        nyc_zip.append(i)
        
nyc_zip = [i[0] for i in nyc_zip if i[0] != 'ZIP']
nyc_zip
```

```python
batteries_final = []
for i in batteries:
    if i['addressZIP'] in nyc_zip:
        batteries_final.append(i)
        
batteries_final
```

```python
import pandas as pd
data = batteries_final

df = pd.DataFrame(data)
df.index.name = 'Index'
df = df.rename(columns={'businessName': 'Sitename', 'addressStreet':'Address', 'addressCity': 'Borough' , 'addressZIP':'Zip' , 'addressLatitude':'Latitude', 'addressLongitude':'Longitude', 'acceptedMaterials':'Category'})
df = df.drop(columns=['accountNumber','addressStateProvince','mainPhone','googlePlaceID', 'hoursMon', 'hoursTue', 'hoursWed','hoursThu','hoursFri','hoursSat','hoursSun','dropOffNotes']) 

df.to_csv('batteries.csv', index=False)

df
```

```python
import folium

batteries_map = folium.Map(location=[df['Latitude'].astype(float).mean(), df['Longitude'].astype(float).mean()], zoom_start=10)

for site_name, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Address'],
        tooltip=row['Sitename']
    ).add_to(batteries_map)

batteries_map
```
