```python
paint = response.json()
```

```python
paint
```

```python
fields_to_keep = ['Name', 'Site_Address', 'City', 'Zip', 'Longitude', 'Latitude']

paint_locations = [
    {field: record['fieldData'][field] for field in fields_to_keep}
    for record in paint['response']['data']
]
```

```python
paint_locations
```

```python
for i in paint_locations:
    i['Category'] = 'Paint'
```

```python
import csv

nyc_zip=[]

with open ("NYC-ZIP.csv","r") as f:
    data = csv.reader(f)
    for i in data:
        nyc_zip.append(i)
```

```python
nyc_zip = [i[0] for i in nyc_zip if i[0] != 'ZIP']
nyc_zip
```

```python
paint_final = []
for i in paint_locations:
    if i['Zip'] in nyc_zip:
        paint_final.append(i)
        
paint_final
```
